I would like to combine two data sets "datasets_90770_209752_insurance" and "insurance_data_set" togather. I think that in business, it is very meaningful to get higher returns at lower costs. From the "insurance_data_set", we can easily get the data distribution of different users purchasing different insurances. First of all, we can use this data set to determine the most profitable group, which is our optimal target customer. Then we can use the distribution in "insurance_claim" to determine the optimal number of target customers in which cities, so that we dont have to invest the same funds in all cities, but can greatly adjust the proportion of investment to those with high profits area.

```python
# combine these two dataset togather.
```

```python
r1 = np.concatenate((data_loader2['age'], data_loader3['age']))
r2 = np.concatenate((data_loader2['charges'], data_loader3['policy_annual_premium']))
r3 = np.concatenate((data_loader2['region'], data_loader3['policy_state']))
newdata = pd.DataFrame({'age':r1,'charges':r2,'region':r3})
print(newdata)
```
